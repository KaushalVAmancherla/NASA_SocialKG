```python
raw_openalex_person_data = read_json_from_local(person_raw_openalex_data_fp)
```

```python
individual_ids = parse_csv_of_individuals(JPL_individual_fp)
```

```python
author_to_works_id = {}
```

```python
# Function to retrieve detailed information about the works of an author
def get_work_ids(url):
    response = requests.get(url)
    if response.status_code == 200:
        #print(len(response.json()['results']))
        works = [metadata["id"].rsplit("/", 1)[-1] for metadata in response.json()['results']]
        
        return works
    else:
        print("Error:", response.status_code)
        return None
```

```python
for person, metadata in raw_openalex_person_data.items():
    i_id = individual_ids[person]
    author_to_works_id[i_id] = get_work_ids(metadata['works_api_url'])
```

```python
works_ids = set()
```

```python
for author,works in author_to_works_id.items():
    works_ids.update(works)
```

```python
len(works_ids)
```

```python
# Function to retrieve detailed information about a works object
def get_works_info(work_id):
    url = f"{base_url}/works/{work_id}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error:", response.status_code)
        return None
```

```python
works_to_response = {}

for work_id in works_ids:
    res = get_works_info(work_id)
    print(work_id)
    
    if res:
        print("success\n")
        works_to_response[work_id] = res
```

```python
write_json_to_local(works_to_response,works_raw_openalex_data_fp)
```
